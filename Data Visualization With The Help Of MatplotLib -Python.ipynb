#### Creating subplots
*It means within one diagram or one canvas we can create multiple plots.

```python
plt.subplot(2,2,1)        #total no of rows=2,no of col=2 and position=1
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

```python
plt.subplot(2,2,1)        
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'go')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)     #we have given 2 rows and 2columns so there will be 4 positions.
plt.plot(x,y,'g*')
```

```python
x=np.arange(1,11)
y=3*x + 5
plt.title('Matplotlib demo')
plt.xlabel('x axis caption')
plt.ylabel('y axis caption')
plt.plot(x,y)
plt.show()
```

```python
#compute the x and y coordinates for points on a sine curve
x = np.arange(0, 4* np.pi,0.1)
y =  np.sin(x)            #sin is a function
plt.title("sine wave form")

#plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
#subplot()
#compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0, 5 * np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

#set up a subplot grid that has height 2 and width 1,
#and set the first such subplot as active.
plt.subplot(2,1,1)

#make the first subplot
plt.plot(x,y_sin,'r--')
plt.title('sine')
#set the second subplot as active, and make the second plot.
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('cosine')

#show the figure
plt.show()
```

#### Bar graph
Bar graph is a graph that represents the categorical data using rectangular bars.

```python
#bar plot
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color = 'g')
plt.title('Bar graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

### Histograms

* Histogram is the graphical representation of frequency of data. They can be used to grasp quick analysis to find the distribution of data.
* A histogram displays numerical data by grouping data into "bins" of equal width. Each bin is plotted as a bar whose height corresponds to how many data points are in that bin. Bins are also sometimes called "intervals", "classes", or "buckets".
* Histograms looks like bar graph but only difference is that the bars are attached to each other,there is no space between the bars.

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a,bins=20)
plt.title('histogram')
```

* In x-axis we will have all values of array 22,87,5,43..and 
* In y-axis it will say the density or count with respect to bins
* by default bins value=10

### Box plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]

#rectangular box plot
plt.boxplot(data, vert=True,patch_artist=True);
```

#### Pie Chart

* A pie chart (or a circle chart) is a circular statistical graphic, which is divided into slices to illustrate numerical proportion. In a pie chart, the arc length of each slice (and consequently its central angle and area) is proportional to the quantity it represents. While it is named for its resemblance to a pie which has been sliced, there are variations on the way it can be presented.

```python
# Data to plot
labels = 'python','C++','Ruby','Java'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.4, 0,0,0)  #explode 1st slice

#plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct = '%1.1f%%',shadow= True)
plt.axis('equal')
plt.show()
```
